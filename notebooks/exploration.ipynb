# ChurnGuard — Telco Customer Churn EDA
Exploratory Data Analysis of the IBM Telco Customer Churn dataset.


```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
sns.set_theme(style="whitegrid")
```

```python
df = pd.read_csv('../data/telco_churn.csv')
print(f"Shape: {df.shape}")
df.head()
```

```python
df.info()
df.describe()
```

```python
df['TotalCharges'] = pd.to_numeric(df['TotalCharges'], errors='coerce')
print(f"Nulls after conversion:\n{df.isnull().sum()[df.isnull().sum() > 0]}")
df['TotalCharges'].fillna(df['TotalCharges'].median(), inplace=True)
```

```python
fig, ax = plt.subplots(figsize=(6, 4))
df['Churn'].value_counts().plot(kind='bar', ax=ax, color=['steelblue', 'tomato'])
ax.set_title('Churn Distribution')
ax.set_xlabel('Churn')
ax.set_ylabel('Count')
plt.tight_layout()
plt.show()
print(df['Churn'].value_counts(normalize=True).round(3))
```

```python
churn_by_contract = df.groupby('Contract')['Churn'].apply(lambda x: (x == 'Yes').mean())
churn_by_contract.plot(kind='bar', figsize=(7, 4), color='coral')
plt.title('Churn Rate by Contract Type')
plt.ylabel('Churn Rate')
plt.xticks(rotation=0)
plt.tight_layout()
plt.show()
```

```python
numeric_df = df.select_dtypes(include=[np.number])
plt.figure(figsize=(10, 7))
sns.heatmap(numeric_df.corr(), annot=True, fmt='.2f', cmap='coolwarm', center=0)
plt.title('Correlation Heatmap — Numeric Features')
plt.tight_layout()
plt.show()
```

```python
fig, ax = plt.subplots(figsize=(10, 4))
for label, color in [('Yes', 'tomato'), ('No', 'steelblue')]:
    df[df['Churn'] == label]['tenure'].hist(ax=ax, bins=30, alpha=0.6, color=color, label=f'Churn={label}')
ax.set_title('Tenure Distribution by Churn')
ax.set_xlabel('Tenure (months)')
ax.legend()
plt.tight_layout()
plt.show()
```

## Key Findings
- **Churn rate**: ~26% of customers churned (class imbalance — consider class_weight or SMOTE)
- **Contract type**: Month-to-month customers churn at much higher rates than 1 or 2-year contracts
- **Tenure**: Customers who churn tend to have shorter tenures — early retention efforts matter
- **TotalCharges**: Had 11 blank entries (converted to NaN, imputed with median)
- **Numeric correlations**: tenure and TotalCharges are highly correlated (both capture customer lifetime)
